### 3. ANALYSE NCBI CSV FILE

```python
# load csv file with NCBI IDs and upload them into server
# on google drive path is somethink like /content/drive/My Drive/..... 

import csv 

# open csv file 
with open('/home/sephyx/Git/DNA_analyser_IBP/example/projekt_02_table.csv', newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter=',') # set delimiter , or \t for tabs ...
        
        next(reader) # if csv file has header like name, id then use next(reader) else delete it
        
        for row in reader: # iterate through rows of csv row[0] = first column row[1] = second ....
           
        # set if circular and tags can be also part of csv then set rows[n]
            API.sequence.ncbi_creator(circular=True, name=row[0], tags=['projekt02'], ncbi_id=row[1])   
```

```python
# to print uploaded sequences use
API.sequence.load_all(filter_tag=['projekt02'])
```

```python
# now save the previous table into variable like this
SEQUENCES = API.sequence.load_all(filter_tag=['projekt02'])
```

```python
# last step is to analyze them
API.g4hunter.analyse_creator(sequence=SEQUENCES, tags=['projekt02'], window_size=25, threshold=1.2)
```

```python
#to show results
API.g4hunter.load_all(filter_tag=['projekt02'])
```

```python
#every table you can also save into csv
API.g4hunter.load_all(filter_tag=['projekt02']).to_csv('/home/sephyx/Git/DNA_analyser_IBP/example/results.csv')
```

```python
#if you want to see detailed result 
G4_RESULTS = API.g4hunter.load_all(filter_tag=['projekt02']) # 1. save table with results

#choose index of analyse you want to see for example 0 by adding iloc[0]
# G4_RESULTS = table .iloc[1] = ROW 1

API.g4hunter.load_results(g4hunter_analyse=G4_RESULTS.iloc[1]) 
```

```python
#to print heatmap of result for example also 1
API.g4hunter.load_heatmap(g4hunter_analyse=G4_RESULTS.iloc[1], segment_count=20, coverage=True)
```

```python
#you can also set different segment counts + change heatmap from coverage to count
API.g4hunter.load_heatmap(g4hunter_analyse=G4_RESULTS.iloc[1], segment_count=30, coverage=False)
```

```python
#last step is to export results into csv
API.g4hunter.export_csv(g4hunter_analyse=G4_RESULTS, out_path='/home/sephyx/Git/DNA_analyser_IBP/example/')
```
